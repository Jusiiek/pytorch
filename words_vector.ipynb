# Exercises

Most popular words

1.1 Find the most related words for woman in Polish

```python
wv.most_similar('kobieta', topn=1)
```

##### nagative samples
1.2 Find the most similar word to the word "woman" that is not similar to the word "man"

```python
wv.most_similar(positive=["kobieta"], negative=["mężczyzna"], topn=1)
```

1.3 Find the most similar word to the word "Pis" that is not similar to the word "Po"

```python
wv.most_similar(positive=['Pis'], negative=['Po'], topn=1)
```

```python
wv.most_similar('Pg', topn=1)
```

##### Search for words that don't match
1.6 In the group "dog cat bison pudding" find the word that does not fit.

```python
wv.doesnt_match(['pies', 'kot', 'żubr', 'budyń'])
```

##### Similarity

1.7 examine the similarity between a cat and a dog

```python
wv.similarity('kot', 'pies')
```

##### Similarity between sentences

Word mover distance - the sum of the minimum distances that each word in a sentence must travel to find a word from the second sentence

image.png

```python
sentence_1 = 'Byłem dzisiaj w kinie'.lower().split()
sentence_2 = 'Poszłam wczoraj do teatru'.lower().split()
```

1.8 Compare the distances between one sentence and another

```python
!pip install POT
```

```python
wv.wmdistance(sentence_1, sentence_2)
```
