### t-SNE visualization

````python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
````

````python
tokens = []
labels = []

for word in list(gwv.index_to_key)[:1000]:
    tokens.append(gwv[word])
    labels.append(word)


tokens = np.array(tokens)
tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
new_values = tsne_model.fit_transform(tokens)

x = []
y = []
for value in new_values:
    x.append(value[0])
    y.append(value[1])

plt.figure(figsize=(16, 16))
for i in range(len(x)):
    plt.scatter(x[i],y[i])
    plt.annotate(labels[i],
                  xy=(x[i], y[i]),
                  xytext=(5, 2),
                  textcoords='offset points',
                  ha='right',
                  va='bottom')
plt.show()
````

### Exercises 2

### Create a bot with a database of sample questions and answers that returns an answer to a query based on the similarity (WMD or flair document embeddings) between the user's query and questions in its database

For example:

Collection of questions and answers:

```python
[
  {"q1" : "a1"},
  {"q2" : "a2"},
  {"q3" : "a3"},
  ...
  {"qn" : "an"},
]

```

Vectorize questions:

```python
[
  {"q1_vecotorized" : "a1"},
  {"q2_vecotorized" : "a2"},
  {"q3_vecotorized" : "a3"},
  ...
  {"qn_vecotorized" : "an"},
]
```

For the user question `q_user` we vectorize it and calculate the similarity with questions in the database

```python
sim("q_user_vectorized","q1_vecotorized"), sim("q_user_vectorized","q2_vecotorized"), ... ("q_user_vectorized""qn_vecotorized")
```

We return the answer to the question from the database that is most similar to the user's question
$$
argmax(sim(q_{user}vectorized,q_{i}vecotorized))
$$

To calculate similarity you can use e.g. https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.cosine.html but then we do argmin because
$$
cosine\ distance = 1 - cosine\ similarity
$$

````python
from flair.embeddings import TransformerDocumentEmbeddings
from flair.data import Sentence
from scipy.spatial.distance import cosine
import numpy as np

dataset = [
    {"question": "What is your name?", "answer": "My name is ChatBOT."},
    {"question": "What is your favorite  color?", "answer": "My favorite color is blue!"},
    {"question": "How are you?", "answer": "I am fine, thanks <3"},
    {"question": "What is your favorite food?", "answer": "I don't have a favorite cuisine, I like any cuisine, as long as it's tasty."},
    {"question": "What do you think about alcohol?", "answer": "Alcohol is very dangerous to health if taken in large quantities, although depending on the type, alcohol can be tasty."},
    {"question": "What are your interests?", "answer": "As an artificial intelligence, I have no particular interests."},
    {"question": "What do you like doing in your spare time?", "answer": "Wait for your question or problem, after all, that's why I'm here."},
    {"question": "Can you write me a song?", "answer": "No, I can't, but I hope I can in the near future."},
    {"question": "Who is your creator?", "answer": "I can not say."},
    {"question": "What should I eat for breakfast?", "answer": "I can recommend you some cereal or a fried egg."},
    {"question": "What should I do when I'm bored?", "answer": "I suggest expanding your knowledge, finding new interests or meeting friends."},
    {"question": "Can you recommend me any series or anime?", "answer": "Sure, from the series I suggest Lucifer, and from the anime Jujutsu Kaisen."},
    {"question": "Can you recommend me some multiplayer games?", "answer": "Sure, the most popular multiplayer games right now are League of Legends, Counter Strike 2 and Valorant."},
    {"question": "Can you recommend me some good cars?", "answer": "According to Google, Toyota sold the most cars in 2023. When it comes to the model, you have to choose yourself."},
    {"question": "What is the weather like today?", "answer": "I can't check the weather right now, but you can use a weather service for that."},
    {"question": "Can you recommend me a good movie?", "answer": "Sure! What genre are you interested in?"},
    {"question": "What is your favorite music genre?", "answer": "As an AI, I don't have personal preferences, but I can appreciate all types of music."},
    {"question": "Who is your favorite musician?", "answer": "I don't have personal preferences, but I can provide information about various musicians."},
    {"question": "What is the capital of France?", "answer": "The capital of France is Paris."},
]

embedding_model = TransformerDocumentEmbeddings('bert-base-uncased')

# vectorize questions
for item in dataset:
    sentence = Sentence(item["question"])
    embedding_model.embed(sentence)
    item["vector"] = sentence.embedding.cpu().numpy()

def find_similar_answer(question):
    sentence = Sentence(question)
    embedding_model.embed(sentence)
    sentence_vector = sentence.embedding.cpu().numpy()

    # calculate similarities
    similarities = []
    for item in dataset:
        similarity = 1 - cosine(sentence_vector, item["vector"])
        similarities.append(similarity)
    
    # returns answer
    most_similar_index = np.argmax(similarities)
    return dataset[most_similar_index]["answer"]


questions = [
    "What should I call you?", 
    "I am bored. Can you recommend me any multiplayer games?",
    "I am hungry, what can you suggest for breakfast?",
    "You know, I'm curious, so I have to ask about your creator. Who's that?",
    "I'm looking for a new genre of music, what can you recommend?"
    ]

for question in questions:
    answer = find_similar_answer(question)
    print(f"User: {question}")
    print(f"Ai: {answer}")
````
